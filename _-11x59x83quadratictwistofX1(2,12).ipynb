```python
###############################################################################
# SageMath code for rank computation of a quadratic twist of X_1(2,12)
# with d = -11 * 59 * 83
#
# This code computes the Mordell–Weil rank of the quadratic twist of the
# elliptic curve corresponding to the modular curve X_1(2,12) by d.
###############################################################################

###############################################################################
# Modular curve X_1(2,12)
###############################################################################

# Define the elliptic curve corresponding to X_1(2,12)
E_2_12 = EllipticCurve([0, -1, 0, 1, 0])
print("Elliptic curve X_1(2,12):")
print(E_2_12)

# Compute its rank
rank_original = E_2_12.rank()
print("Rank of X_1(2,12):", rank_original)

###############################################################################
# Quadratic twist by d = -11 * 59 * 83
###############################################################################

# Define d
d = -11 * 59 * 83

# Define the quadratic twist
E_twist_d = E_2_12.quadratic_twist(d)
print(f"\nQuadratic twist of X_1(2,12) by d = {d}:")
print(E_twist_d)

# Compute the Mordell–Weil rank of the twisted curve
rank_twist_d = E_twist_d.rank()
print(f"Rank of the quadratic twist by d = {d}:", rank_twist_d)


###############################################################################
# Observation:
# The original curve has rank 0, while this quadratic twist has rank 2.
###############################################################################
```
