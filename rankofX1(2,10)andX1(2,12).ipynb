```python
###############################################################################
# SageMath code for rank computations of modular curves X_1(2,10) and X_1(2,12)
#
# This code computes the Mordell–Weil rank of the elliptic curves
# corresponding to the modular curves X_1(2,10) and X_1(2,12).
#
# Sage version tested: SageMath >= 9.0
###############################################################################

###############################################################################
# Modular curve X_1(2,10)
###############################################################################

# Define the elliptic curve corresponding to X_1(2,10)
E_2_10 = EllipticCurve([0, 1, 0, -1, 0])

# Display basic information
print("Elliptic curve X_1(2,10):")
print(E_2_10)

# Compute the Mordell–Weil rank
rank_2_10 = E_2_10.rank()

# Output the rank
print("Rank of X_1(2,10):", rank_2_10)

###############################################################################
# Modular curve X_1(2,12)
###############################################################################

# Define the elliptic curve corresponding to X_1(2,12)
E_2_12 = EllipticCurve([0, -1, 0, 1, 0])

# Display basic information
print("\nElliptic curve X_1(2,12):")
print(E_2_12)

# Compute the Mordell–Weil rank
rank_2_12 = E_2_12.rank()

# Output the rank
print("Rank of X_1(2,12):", rank_2_12)

###############################################################################
# Conclusion:
# In both cases, Sage verifies that the Mordell–Weil rank is equal to 0.
###############################################################################
```
