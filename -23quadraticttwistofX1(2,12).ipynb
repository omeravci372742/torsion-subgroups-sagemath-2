```python
###############################################################################
# SageMath code for rank computation of a quadratic twist of X_1(2,12)
#
# This code computes the Mordell–Weil rank of the quadratic twist of the
# elliptic curve corresponding to the modular curve X_1(2,12) by -23.
###############################################################################

###############################################################################
# Modular curve X_1(2,12)
###############################################################################

# Define the elliptic curve corresponding to X_1(2,12)
E_2_12 = EllipticCurve([0, -1, 0, 1, 0])
print("Elliptic curve X_1(2,12):")
print(E_2_12)

# Compute its rank
rank_original = E_2_12.rank()
print("Rank of X_1(2,12):", rank_original)

###############################################################################
# Quadratic twist by -23
###############################################################################

# Define the quadratic twist
D = -23
E_twist = E_2_12.quadratic_twist(D)
print(f"\nQuadratic twist of X_1(2,12) by {D}:")
print(E_twist)

# Compute the Mordell–Weil rank of the twisted curve
rank_twist = E_twist.rank()
print(f"Rank of the quadratic twist by {D}:", rank_twist)

###############################################################################
# Observation:
# The original curve has rank 0, while its -23 quadratic twist has rank 1.
###############################################################################
```
